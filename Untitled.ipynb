```python
import pandas as pd
import numpy as np
def SEIRmodel(Tspan, initial,para):
#Tspan运行的年份
#initial=np.array([S,E,I,R]) 初值
#para为对应每一年模型的参数， 这里假设所有参数都依赖于时间， 如果常数的话，每年都一样
    columns=['S','E','I','R']
    index=range(Tspan)
    sol=pd.DataFrame(columns=columns,index=index)#定义一个初始的dataframe记录每年SEIR
    S=initial[0];E=initial[1];I=initial[2];R=initial[3];
    for i in range(Tspan):
        sol.loc[i,'S']=S+para.loc[i,'birth'] -para.loc[i,'death']*S - para.loc[i,'infection']*S* I
        sol.loc[i,'E']=E+ para.loc[i,'infection']*S* I -(para.loc[i,'death']+para.loc[i,'k'])*E 
        sol.loc[i,'I']=I+para.loc[i,'k']*E-(para.loc[i,'recover']+para.loc[i,'death'])*I
        sol.loc[i,'R']=R+para.loc[i,'recover']*I-para.loc[i,'death']*R
        S=sol.loc[i,'S']
        E=sol.loc[i,'E']
        I=sol.loc[i,'I']
        R=sol.loc[i,'R']
    return sol

Tspan=len([186]*500+[186,142,168,161,162,139,68,73,183,224,182,187,191,168,202,236,211,205,202,202,184,158,148,148,154,
         153,145,151,153,154,145,157,187,212,214,223,214,172,159,141,138,130])
columns=['birth','death','k','infection','recover']
para=pd.DataFrame(columns=columns,index=range(Tspan))
a = [186]*500+[186,142,168,161,162,139,68,73,183,224,182,187,191,168,202,236,211,205,202,202,184,158,148,148,154,
         153,145,151,153,154,145,157,187,212,214,223,214,172,159,141,138,130]
para['birth'] = [i * 10000 for i in a]
#para['birth']=1   
b = [13.32] * 500 + [13.32,11.75,14,11.8,12.69,14.1,39.56,10.2, 8.04,9.43,10.61,8.45,8.24,8.59,7.91, 7.11,7.61,7.46,7.17,
            7.27,7.44,7.66,7.18,7.04,6.3,6.35,6.32,6.57, 6.21,6.3,6.26,6.13,6.44,6.32,5.83,5.76,6.52,6.63,6.99,6.35,
            6.34,6.28]
para['death'] = [i / 1000 for i in b]
#para['death']=0.1
para['k']=0.1
para['infection']=0.05/10000000
para['recover']=0.1


N = 100 # 总人口
I0 =1 # 期初感染数
E0 = 0 # 期初潜伏数
RM0 = 0 # 期初移除数
S0 = N - I0 - RM0 # 期初易感人数
initial=np.array([S0, E0, I0, RM0])	
sol=SEIRmodel(Tspan, initial,para)
print(sol)


import matplotlib.pyplot as plt
sol.plot()
```
